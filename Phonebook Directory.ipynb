```python
class Node:
    def __init__(self,name,no):
        self.next = None
        self.prev = None
        self.name = name
        self.no = no

class Llist:
    def __init__(self):
        self.head = None
        self.tail = None

    def add(self,x,y):
        n = Node(x,y)
        if self.head is None:
            self.head = n
            self.tail = self.head
            return
        n.prev = self.tail
        self.tail.next = n
        self.tail = n

    def show(self):
        print("Contacts saved are:")
        print("--------------")
        tmp = self.head
        while(tmp):
            print(tmp.name, end = ": ")
            print(tmp.no, end = "")
            print()
            tmp = tmp.next
        print("______________")
    
    def find(self,k):
        tmp = self.head
        while(tmp):
            if (tmp.name).lower() == k.lower():
                print(k,":",tmp.no)
                return
            tmp = tmp.next
        print("No contact found ")

    def delete(self,name):
        tmp = self.head
        if self.head is None:
            print("No contact found")
            print()
            return
        if (self.head.name).lower() == name.lower():
            self.head = tmp.next
            print(name + " deleted")
            print()
            return
        if (self.tail.name).lower() == name.lower():
            self.tail = self.tail.prev
            self.tail.next = None
            print(name,"deleted")
            print()
            return
        tmp = tmp.next
        while(tmp):
            if (tmp.name).lower() == name.lower():
                tmp.prev.next = tmp.next
                tmp.next.prev = tmp.prev
                print(name,"deleted")
                print()
                return
            tmp = tmp.next
        print("No contact found")
flag = 0            
l = Llist()
print("Welcome to phonebook")
while(flag == 0):
    print()
    print("Phonebook")
    print("---------------")
    print("Enter key to perform the operation ")
    print("Show Phonebook        : 1")
    print("Add new contact       : 2")
    print("Find contact number   : 3")
    print("Delete contact number : 4")
    print("Close phonebook       : 5")
    print()
    x = int(input("Key: "))
    while(x < 1 or x > 5):
        x = int(input("Key: "))
    if x == 1:
        l.show()
    elif x == 2:
        name = input("Enter name: ")
        number = input("Enter number: ")
        while(len(number) == 0):
            number = input("Enter number: ")
        l.add(name,number)
    elif x == 3:
        a = input("Enter name: ")
        l.find(a)
    elif x == 4:
        name = input("Enter name: ")
        l.delete(name)
    else:
        flag = 1
```
